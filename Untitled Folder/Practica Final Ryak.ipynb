```python
myClient = riak.RiakClient()
myClient.ping()
```

```python
BUCKET_JUEGO = 'juego'
BUCKET_CONSOLA = 'consola'
BUCKET_RATINGS = 'ratings'

juego = myClient.bucket(BUCKET_JUEGO)
consola = myClient.bucket(BUCKET_CONSOLA)
ratings = myClient.bucket(BUCKET_RATINGS)
```

```python
def insert_consola(consola):
    count = consola.get(consola).data
    if count == None:
        count = 1
    else:
        count = count + 1

    consola.new(consola, data = count).store()
```

```python
def insert_juego(juegos_json):
    juegos = juego.new(juegos_json["Name"], juegos_json)    
    juegos.store()
```
